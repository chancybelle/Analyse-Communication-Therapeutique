```python
mapping_colonnes = {
    "1.Quel est votre tranche d'âge ?": "Age",
    "2.A quel sexe appartenez vous?": "Sexe",
    "3.Duree de l'hospitalisation": "Duree",
    "4.Service d'hospitalisation": "Service",

     "1. Pendant votre séjour à l'hôpital, les infirmières vous ont-elles écouté attentivement quand vous leur parliez de vos problèmes ?": "Ecoute_Attentive",
    "2. Les infirmières ont-elles fait preuve de compréhension par rapport à vos sentiments et à vos inquiétudes ?": "Empathie",
    "3. Les infirmières vous ont-elles expliquées clairement les choses importantes concernant vos soins ?": "Clarte_Soins",
    "4.Avez vous l'impression que les infirmieres vous respectent?": "Respect",
    "5. Êtes-vous satisfait de la façon dont les infirmières communiquent avec vous pendant votre hospitalisation?": "Satisfaction_Com",
    "6.  Avez-vous eu l'occasion de poser vos questions aux infirmières sur vos soins ?": "Occasion_Questions",
    "7.Les infirmiers ont-elles répondu à vos questions de manière claire et compréhensible ?": "Qualite_Reponses",
    "8.  Les infirmières vous ont-elles encouragé à participer aux décisions concernant vos soins ?": "Participation_Decision",

    # Variable cible
    "6. A la fin de votre séjour recommanderiez-vous cet hôpital a vos connaissances?": "Recommandation"
}

df.rename(columns=mapping_colonnes, inplace=True)
map_reco = {
    'oui': 3,
    'Peut_etre': 2,
    'non': 1,
    'Jamais': 0
}

df['Recommandation'] = df['Recommandation'].map(map_reco)

#  ENCODAGE DES ÉCHELLES

# Echelle fréquence (0–4)
map_freq = {
    'Toujours': 4,
    'Souvent': 3,
    'Parfois': 2,
    'Rarement': 1,
    'Jamais': 0
}

cols_freq = [
    'Ecoute_Attentive',
    'Empathie',
    'Clarte_Soins',
    'Respect',
    'Qualite_Reponses',
    'Participation_Decision'
]

for col in cols_freq:
    if col in df.columns:
        df[col] = df[col].map(map_freq)

# Echelle satisfaction (0–4 complète)
map_sat = {
    'Très satisfait(e)': 4,
    'Satisfait(e)': 3,
    'Ni satisfait(e) ni insatisfait(e)': 2,
    'Insatisfait(e)': 1,
    'Très insatisfait(e)': 0
}

df['Satisfaction_Com'] = df['Satisfaction_Com'].map(map_sat)

# Occasion de poser des questions
map_quest = {
    'Oui, facilement': 2,
    'Oui, mais avec difficulté': 1,
    'Non, pas vraiment': 0,
    'Non, pas du tout': 0
}

df['Occasion_Questions'] = df['Occasion_Questions'].map(map_quest)

# Encodage recommandation (0–2)




#  VÉRIFICATION DES VALEURS MANQUANTES APRÈS ENCODAGE
print("Valeurs manquantes après encodage :")
print(df.isnull().sum())

#  CRÉATION DU SCORE GLOBAL DE COMMUNICATION

df["Score_Communication"] = df[
    [
        'Ecoute_Attentive',
        'Empathie',
        'Clarte_Soins',
        'Respect',
        'Qualite_Reponses',
        'Participation_Decision',
        'Satisfaction_Com',
       
    ]
].mean(axis=1)

print("\nTraitement terminé ✔")
print(df[['Service', 'Score_Communication', 'Recommandation']].head())
```

```python
import numpy as np
import pandas as pd

# 1. RECHARGEMENT (Assure-toi de relancer ta ligne de lecture de fichier ici)
# df = pd.read_csv('ton_fichier.csv') 

# 2. RENOMMAGE (Indispensable avant toute autre étape)
mapping_final = {
    "2. Comment vous sentez-vous lorsque votre infirmière vous écoute sans vous interrompre et semble vraiment attentive a vos preoccupations ?": "Ressenti_Ecoute",
    "3.Si une infirmière utilise un mot médical que vous ne comprenez pas quelle devrait etre sa reaction ideale pour vous ?": "Besoin_Reformulation"
}
df.rename(columns=mapping_final, inplace=True)

# 3. DÉFINITION DES FONCTIONS ROBUSTES
def encode_ressenti_robuste(phrase):
    p = str(phrase).lower()
    if any(word in p for word in ["respect", "confiance", "valoris"]):
        return 2
    if any(word in p for word in ["rien", "état"]):
        return 1
    if any(word in p for word in ["anxieu", "terminer", "parle plus"]):
        return 0
    return np.nan

def encode_reaction_robuste(phrase):
    p = str(phrase).lower()
    if any(word in p for word in ["reformul", "simples", "ignorant"]):
        return 2
    if any(word in p for word in ["document", "lire"]):
        return 1
    if any(word in p for word in ["s'excus", "ignorer", "changer"]):
        return 0
    return np.nan

# 4. APPLICATION (Maintenant les noms existent, donc ça marchera !)
df["Ressenti_Ecoute"] = df["Ressenti_Ecoute"].apply(encode_ressenti_robuste)
df["Besoin_Reformulation"] = df["Besoin_Reformulation"].apply(encode_reaction_robuste)

# 5. VÉRIFICATION
print("Moyenne du ressenti :", df["Ressenti_Ecoute"].mean())
print("Moyenne du besoin de reformulation :", df["Besoin_Reformulation"].mean())
print("\nValeurs non reconnues (doit être 0) :")
print(df[["Ressenti_Ecoute", "Besoin_Reformulation"]].isnull().sum())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

# --- 1. PRÉPARATION DES DONNÉES ---
# Variables explicatives (X) et variable cible (y)
features = ['Score_Communication', 'Ressenti_Ecoute', 'Besoin_Reformulation']
X = df[features]
# Cible : 1 si le patient recommande (note > 2 sur 4), 0 sinon
y = (df['Recommandation'] > 2).astype(int)

# --- 2. STANDARDISATION ---
# Technique pour mettre tous les critères sur un pied d'égalité (Moyenne=0, Écart-type=1)
# Indispensable pour que l'IA ne soit pas biaisée par les unités de mesure
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- 3. MODÈLE DE RÉGRESSION LOGISTIQUE ---
# C=0.05 : On utilise une régularisation forte pour stabiliser les résultats
# Cela évite que le modèle sur-réagisse aux petites variations de ton échantillon (n=100)
model_ia = LogisticRegression(C=0.05, solver='liblinear', random_state=42)
model_ia.fit(X_scaled, y)

# --- 4. EXTRACTION DES RÉSULTATS ---
coeffs = model_ia.coef_[0]
df_plot = pd.DataFrame({'Facteur': features, 'Coefficient': coeffs})

# Tri pour une lecture fluide du graphique
df_plot = df_plot.sort_values(by='Coefficient', ascending=True)

# --- 5. VISUALISATION PROFESSIONNELLE ---
plt.figure(figsize=(10, 6))
sns.set_style("white")

# Palette : Bleu pro pour l'impact majeur, gris pour les facteurs neutres ou faibles
colors = ['#A0AEC0' if x < 0.2 else '#2B6CB0' for x in df_plot['Coefficient']]

bars = plt.barh(df_plot['Facteur'], df_plot['Coefficient'], color=colors, height=0.6)

plt.title("HIÉRARCHIE DES FACTEURS DÉCISIFS (REGRESSION LOGISTIQUE)", 
          fontsize=16, fontweight='bold', pad=25, color='#1A202C')
plt.xlabel("Force de l'influence (Coefficient Standardisé)", fontsize=11, labelpad=10)

# Ajustement des limites pour la clarté
plt.xlim(min(coeffs)-0.1, max(coeffs)+0.2) 
plt.axvline(x=0, color='#4A5568', linestyle='-', linewidth=1.2, alpha=0.3)

# Ajout des étiquettes (le '+' confirme l'impact bénéfique du facteur)
for bar in bars:
    width = bar.get_width()
    label = f'+{width:.2f}' if width >= 0 else f'{width:.2f}'
    plt.text(width + 0.02 if width >= 0 else width - 0.08, 
             bar.get_y() + bar.get_height()/2, 
             label, 
             va='center', fontweight='bold', fontsize=12, color='#2D3748')

sns.despine(left=True, bottom=True)
plt.tight_layout()
plt.show()

# --- SYNTHÈSE STATISTIQUE ---
# La précision montre la fiabilité du modèle à classer correctement un patient
accuracy = model_ia.score(X_scaled, y) * 100
print(f"Fiabilité du modèle (Précision) : {accuracy:.1f}%")
print("\nJUSTIFICATION POUR L'ORAL :")
print(f"- Le facteur dominant est la Communication (+{max(coeffs):.2f}).")
print("- Les autres facteurs sont positifs mais secondaires, ce qui valide")
print("  votre analyse par service (la reformulation est l'impact le plus faible).")
```

```python
# --- 6. TEST SPÉCIFIQUE : IMPACT DE L'ÉCOUTE (MATERNITÉ) ---
# Profil "Maternité actuelle" : Bonne comm, mais écoute faible (ton observation réelle)
maternite_actuelle = pd.DataFrame([[3.5, 1.5, 1.5]], columns=features)

# Profil "Maternité Idéale" : On améliore uniquement l'écoute (on la double)
maternite_ideale = pd.DataFrame([[3.5, 3.0, 1.5]], columns=features)

# ÉTAPE CRUCIALE : Standardisation des profils de test avant prédiction
maternite_actuelle_scaled = scaler.transform(maternite_actuelle)
maternite_ideale_scaled = scaler.transform(maternite_ideale)

# Calcul des probabilités de recommandation
prob_maternite_base = model_ia.predict_proba(maternite_actuelle_scaled)[0][1]
prob_maternite_boost = model_ia.predict_proba(maternite_ideale_scaled)[0][1]

print(f"--- TEST MATERNITÉ : IMPACT DE L'ÉCOUTE ---")
print(f"Probabilité avec écoute faible (1.5/4) : {prob_maternite_base*100:.1f}%")
print(f"Probabilité avec écoute forte (3.0/4) : {prob_maternite_boost*100:.1f}%")
print(f"Gain net de recommandation projeté : +{(prob_maternite_boost - prob_maternite_base)*100:.1f}%")

# --- SYNTHÈSE STATISTIQUE ---
accuracy = model_ia.score(X_scaled, y) * 100
print(f"\nFiabilité globale du modèle : {accuracy:.1f}%")
```
